```python
#importing libraries
import numpy as np
import pandas as pd
from pandas import DataFrame,read_csv

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Passing Mexico city temperature csv file into a DataFrame
#file = pd.read_csv("/Users/polly/Documents/Ironhack/Dronostic-esky/México/Temperatura Ciudad de México.csv")

#Pasé el archivo CSV de Ciudad de México a un marco de datos PANDAS y 
#creé una nueva columna que contiene mis promedios móviles.

file = r'Temperatura_CDMX.csv'
Temperatura_CDMX= pd.read_csv(file)


Temperatura_CDMX['moving_temp'] = Temperatura_CDMX.avg_temp.rolling(window=20).mean() 

Temperatura_CDMX

```

```python
#Crear un gráfico de líneas para los promedios móviles de las temperaturas de la Ciudad de México.

plt.plot(Temperatura_CDMX.Año,Temperatura_CDMX.moving_temp)
plt.title('Ciudad de Mexico Temperatures Moving Averages')
plt.ylabel('Temperatures')
plt.xlabel('Year')
plt.show()
```

```python
#Pasé un archivo csv de temperatura global a un DataFrame

file=r'GlobalTemperature.csv'
globaltemp = pd.read_csv(file)

# Agregar las medias móviles de las temperaturas globales

globaltemp['moving_temp'] = globaltemp.avg_temp.rolling(window=20).mean() 

globaltemp
```

```python
#Crear un gráfico de líneas para las medias móviles de las temperaturas globales.

plt.plot(globaltemp.year,globaltemp.moving_temp)
plt.title('Global Temperatures Moving Averages')
plt.ylabel('Temperatures')
plt.xlabel('Year')
plt.show()
```

```python
#Creé un gráfico de líneas que combina los puntos de datos de los promedios móviles de las 
#temperaturas globales y de La ciudad de México en un solo gráfico.

fig, ax = plt.subplots()
ax2 = ax.twinx()

Temperatura_CDMX.plot(x='Año', y='moving_temp', ax=ax, color='purple', ls='--')
globaltemp.plot(x='year', y='moving_temp', ax=ax2, color='green')
ax2.legend().set_visible(False)
ax.legend().set_visible(False)
ax2.set_ylabel('Global Temperatures',color='green',fontsize=14)
ax.set_ylabel('Mexico Temperatures',color='purple', fontsize=14)


```
