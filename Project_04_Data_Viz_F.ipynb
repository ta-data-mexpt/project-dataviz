```python
argentina_list = pd.read_csv('Argentina/listings.csv')
brasil_list = pd.read_csv('Brasil/listings.csv')
chile_list = pd.read_csv('Chile/listings.csv')
mexico_list = pd.read_csv('Mexico/listings.csv')
```

```python
columnas_utiles = ['id', 'name', 'description','neighborhood_overview', 'host_id',
                   'host_name', 'host_since', 'host_acceptance_rate','host_is_superhost',
                   'neighbourhood_cleansed', 'latitude','longitude','room_type', 'accommodates',
                   'bathrooms_text', 'bedrooms', 'beds', 'amenities', 'price','minimum_nights',
                   'maximum_nights', 'number_of_reviews', 'first_review','last_review',
                   'review_scores_rating', 'instant_bookable','reviews_per_month']
```

```python
argentina_list = argentina_list[columnas_utiles]
brasil_list = brasil_list[columnas_utiles]
chile_list = chile_list[columnas_utiles]
mexico_list = mexico_list[columnas_utiles]
```

```python
columnas_arg = argentina_list.isnull().sum()[argentina_list.isnull().sum()>0]
columnas_arg
```

```python
columnas_bra = brasil_list.isnull().sum()[brasil_list.isnull().sum()>0]
columnas_bra
```

```python
columnas_chi = chile_list.isnull().sum()[chile_list.isnull().sum()>0]
columnas_chi
```

```python
columnas_mex = mexico_list.isnull().sum()[mexico_list.isnull().sum()>0]
columnas_mex
```

```python
reviews_nulos_arg = argentina_list[argentina_list['review_scores_rating'].isnull()==True]
reviews_nulos_arg[['review_scores_rating','reviews_per_month']].isnull().sum()
argentina_list2 = argentina_list.drop(reviews_nulos_arg.index , axis=0)
print('registros iniciales Argentina: ', len(argentina_list))
print('registros finales   Argentina: ', len(argentina_list2))
print('registos totales   eliminados: ',(len(argentina_list)-len(argentina_list2)))
```

```python
reviews_nulos_bra = brasil_list[brasil_list['review_scores_rating'].isnull()==True]
reviews_nulos_bra[['review_scores_rating','reviews_per_month']].isnull().sum()
brasil_list2 = brasil_list.drop(reviews_nulos_bra.index , axis=0)
print('registros iniciales   Brasil: ', len(brasil_list))
print('registros finales     Brasil: ', len(brasil_list2))
print('registros totales eliminados: ',(len(brasil_list)-len(brasil_list2)))
```

```python
reviews_nulos_chi = chile_list[chile_list['review_scores_rating'].isnull()==True]
reviews_nulos_chi[['review_scores_rating','reviews_per_month']].isnull().sum()
chile_list2 = chile_list.drop(reviews_nulos_chi.index , axis=0)
print('registros iniciales    Chile: ', len(chile_list))
print('registros finales      Chile: ', len(chile_list2))
print('registros totales eliminados: ', (len(chile_list)-len(chile_list2)))
```

```python
reviews_nulos_mex = mexico_list[mexico_list['review_scores_rating'].isnull()==True]
reviews_nulos_mex[['review_scores_rating','reviews_per_month']].isnull().sum()
mexico_list2 = mexico_list.drop(reviews_nulos_mex.index , axis=0)
print('registros iniciales   Mexico: ', len(mexico_list))
print('registros finales     Mexico: ', len(mexico_list2))
print('registros totales eliminados: ', (len(mexico_list)-len(mexico_list2)))
```

```python
argentina_list2['pais'] = 'Argentina'
brasil_list2['pais'] = 'Brasil'
chile_list2['pais'] = 'Chile'
mexico_list2['pais'] = 'Mexico'

argentina_list2['ciudad'] = 'Buenos Aires'
brasil_list2['ciudad'] = 'Rio de Janeiro'
chile_list2['ciudad'] = 'Santiago de Chile'
mexico_list2['ciudad'] = 'CDMX'
```

```python
argentina_list2['price']
```

```python
argentina_list2['price'] = argentina_list['price'].str.replace('$','').str.replace(',','').astype('float64')
argentina_list2['price']
```

```python
brasil_list2['price'] = brasil_list2['price'].str.replace('$','').str.replace(',','').astype('float64')
chile_list2['price'] = chile_list2['price'].str.replace('$','').str.replace(',','').astype('float64')
mexico_list2['price'] = mexico_list2['price'].str.replace('$','').str.replace(',','').astype('float64')
```

```python
argentina_list2['Precio_MXN'] = (argentina_list2['price']*0.22)
print(argentina_list2['price'])
```

```python
argentina_list2['Precio_MXN']
```

```python
brasil_list2['Precio_MXN'] = (brasil_list2['price']*4.06)
chile_list2['Precio_MXN'] = (chile_list2['price']*0.028)
mexico_list2['Precio_MXN'] = (mexico_list2['price']*1)
```

```python
todos_paises = pd.concat([argentina_list2,brasil_list2,chile_list2,mexico_list2], axis= 0)
```
