```python
url = 'https://www.federalreserve.gov/paymentsystems/coin_currcircvalue.htm'
html = requests.get(url).content
soup=BeautifulSoup(html,'lxml')
soup.title.text
```

```python
# Get column names
index = soup.find_all('tr',{'class':'header'})
list_col = soup.find_all('th',{'scope':'col'})
nomcol = [list.text.replace('\n','').replace('\t','') for list in list_col]

#Get data of the table and header of the table
yearlist = soup.find_all('td',{'class':'left'})
usdlist = soup.find_all('td',{'class':'left2'})
year = [list.text.replace('\n','').replace('\t','') for list in yearlist]
usd = [list.text.replace('\n','').replace('\t','') for list in usdlist]

#Form the correct shape for the date
npyear= np.array(year)
npusd= np.array(usd)
npyearc=npyear.reshape(21,1)
npusdc=npusd.reshape(21,9)
npyear = [list.replace(',','') for list in npyear]

#Create DataFrame
usd_cry_df = pd.DataFrame(npusdc,index=npyear,columns=nomcol[1:])


#Create DataFrame in float type
usd_cry_df2=pd.DataFrame()
usd_cry_df2 = usd_cry_df[usd_cry_df.columns[1:]].replace('[\$,]', '', regex=True).astype(float)
#print(usd_cry_df2)

#Plot Data Frame
usd_cry_df.index = pd.to_datetime(usd_cry_df.index)
usd_cry_df2.index = pd.to_datetime(usd_cry_df2.index)
usd_cry_df2['Total'].plot()
plt.title(soup.title.text)
plt.show()
print(usd_cry_df)
```
