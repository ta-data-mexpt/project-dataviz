# ¿Afecta el idioma?

```python
# Number of books in each language

all_languages = books[["language_code"]].value_counts()

english = all_languages["eng"] + all_languages["en-US"] + all_languages["en-GB"] + all_languages["en-CA"]

all_languages.replace(all_languages["eng"], english, inplace = True)
all_languages.drop(labels = ["en-US", "en-GB", "en-CA"], inplace = True)

print(all_languages)
print("")

languages = {"English":"eng", "Spanish":"spa", "French":"fre", "German":"ger", "Japanese":"jpn"}

for language in languages.items():
    proportion = round(((all_languages[language[1]]/all_languages.sum())*100),2)
    print(f"The percentage of books in {language[0]} is {proportion}%")
print("")    

all_languages.drop(labels = ["eng"], inplace = True)

print(all_languages)
print("")

new_languages = {"Spanish":"spa", "French":"fre", "German":"ger", "Japanese":"jpn", "Multilanguage": "mul", "Chinese": "zho",
                "Ancient Greek": "grc", "Portuguese": "por", "Italian": "ita", "Latin": "lat", "Middle English": "enm", 
                 "Swedish": "swe", "Russian": "rus", "Gaelic": "gla", "Welsh": "wel", "Turkish": "tur", "Serbian": "srp",
                "Norwegian": "nor", "Galician": "glg", "Dutch": "nl", "Malay": "msa", "Arabic": "ara", "Aleut": "ale"}

for new_language in new_languages.items():
    new_proportion = round(((all_languages[new_language[1]]/all_languages.sum())*100),2)
    print(f"If we take English out, the percentage of books in {new_language[0]} is {new_proportion}%")
```

```python
all_languages_df = all_languages.to_frame().reset_index()
all_languages_df.rename(columns = {0: "number_of_books"}, inplace = True)

all_languages_df.head()
```

```python
fig = px.bar(all_languages_df, x = "language_code", y = "number_of_books")
fig.show()
```

```python
real_books = books[books.num_pages >= 50]
print(real_books["num_pages"].mode())
real_books["num_pages"].describe()
```

```python
rated_books = real_books[real_books.ratings_count >= 50]
```

```python
spanish_rated_books = rated_books[rated_books["language_code"] == "spa"]
```

```python
french_rated_books = rated_books[rated_books["language_code"] == "fre"]
```

```python
german_rated_books = rated_books[rated_books["language_code"] == "ger"]
```

```python
japanese_rated_books = rated_books[rated_books["language_code"] == "jpn"]
```

```python
english_rated_books = rated_books[(rated_books["language_code"] == "eng") | (rated_books["language_code"] == "en-US")
                                 | (rated_books["language_code"] == "en-GB") | (rated_books["language_code"] == "en-CA")]
```

```python
ratings_means = pd.Series(data = {"English": english_rated_books["average_rating"].mean(), "Spanish": spanish_rated_books["average_rating"].mean(), 
                                  "French": french_rated_books["average_rating"].mean(), "German": german_rated_books["average_rating"].mean(), 
                                  "Japanese": japanese_rated_books["average_rating"].mean()})

ratings_maxs = pd.Series(data = {"English": english_rated_books["average_rating"].max(), "Spanish": spanish_rated_books["average_rating"].max(), 
                                  "French": french_rated_books["average_rating"].max(), "German": german_rated_books["average_rating"].max(), 
                                  "Japanese": japanese_rated_books["average_rating"].max()})

ratings_mins = pd.Series(data = {"English": english_rated_books["average_rating"].min(), "Spanish": spanish_rated_books["average_rating"].min(), 
                                  "French": french_rated_books["average_rating"].min(), "German": german_rated_books["average_rating"].min(), 
                                  "Japanese": japanese_rated_books["average_rating"].min()})

ratings_df = pd.DataFrame({"ratings_means": ratings_means, "ratings_maxs": ratings_maxs, "ratings_mins": ratings_mins})
```

```python
ratings_df.head()
```

```python
languages=["English", "Spanish", "French", "German", "Japanese"]

fig = go.Figure(data=[
    go.Bar(name='ratings_mean', x=languages, y= ratings_df["ratings_means"]),
    go.Bar(name='ratings_max', x=languages, y= ratings_df["ratings_maxs"]),
    go.Bar(name='ratings_min', x=languages, y= ratings_df["ratings_mins"])
])
# Change the bar mode
fig.update_layout(title="Ratings comparison by language", xaxis_title="Languages", yaxis_title="Rating", barmode='group')
fig.show()
```

# ¿Afecta el año de publicación?

```python
rated_books = real_books[real_books.ratings_count >= 50]
rated_books.head()
```

```python
rated_books["publication_year"] = rated_books.publication_date.str[-4:].astype(int)
```

```python
years_sample = pd.DataFrame(rated_books[["publication_year"]].value_counts().reset_index().values, columns = ["Year", "total_books"])

years_sample.head()
```

```python
years_sample["percentage"] = (years_sample["total_books"]/8903)*100

years_sample["sample_books"] = years_sample["total_books"]*(years_sample["percentage"]/100)

years_sample["sample_books"] = years_sample["sample_books"].apply(lambda x: int(x))

years_sample_no_zero = years_sample[years_sample["sample_books"] > 0]
```

```python
year_samples_new = {}

for year in years_sample_no_zero["Year"].values:
    books_for_year_new = rated_books[rated_books["publication_year"] == year]
    year_samples_new[f"{year}"] = books_for_year_new
```

```python
books_sample_year_new = pd.concat(year_samples_new.values())
```

```python
fig = px.scatter(x=books_sample_year_new["publication_year"], y= books_sample_year_new["average_rating"], opacity = 0.3, 
                 title = "Average rating VS. Publication Year")
fig.show()
```

# ¿Afecta el número de libros publicados por cada autor?

```python
rated_books = real_books[real_books.ratings_count >= 50]

grouped_authors = rated_books.groupby("authors")

grouped_multibook_authors_df = grouped_authors.filter(lambda x: len(x) > 1)

authors_ratings_count_mean = grouped_multibook_authors_df.groupby("authors")["ratings_count"].mean()

authors_means = grouped_multibook_authors_df.groupby("authors", as_index = False)["average_rating"].mean()

authors_books = grouped_multibook_authors_df.groupby("authors", as_index = False).size()

authors_books.rename(columns = {"size": "total_books"}, inplace = True)
```

```python
authors_ratings_count_mean_df = authors_ratings_count_mean.to_frame()
```

```python
authors_ratings_count_mean_df.reset_index(level=0)
```

```python
final_authors = pd.merge(authors_books, authors_ratings_count_mean_df, on = "authors")

final_authors.describe()

final_authors.head()
```

```python
fig = px.scatter(x= final_authors["total_books"], y= final_authors["ratings_count"], opacity = 0.7, 
                 title = "Number of books VS. Average number of reviews")
fig.show()
```

# ¿Afecta el número de reviews medio de cada autor con más de un libro?

```python
final_authors = pd.merge(authors_books, authors_means, on = "authors")

final_authors.describe()

final_authors.head()
```

```python
fig = px.scatter(x= final_authors["total_books"], y= final_authors["average_rating"], opacity = 0.6, 
                 title = "Number of books VS. Average rating")
fig.show()
```
